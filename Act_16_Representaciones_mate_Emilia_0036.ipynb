#### problema 1
 La suma de un número más 9 es igual a 21Paso 1: Análisis y resolución matemáticaIdentificar la incógnita: El problema nos pide un número que no conocemos. Lo representaremos con la letra $x$.Plantear la ecuación: * "La suma de un número ($x$) más el número 9" se escribe: $x + 9$."Es igual a 21" completa nuestra igualdad:$$x + 9 = 21$$Despejar la variable: Para dejar a la $x$ sola, debemos mover el 9 al otro lado de la igualdad. Como el 9 está sumando, pasa realizando la operación contraria (restando):$x = 21 - 9$Resolver:$x = 12$Solución: El número buscado es 12.

```python
# Problema 1: Encontrar un número x tal que x + 9 = 21
# Definimos los valores conocidos
# resultado_total = 21
# numero_sumado = 9
resultado_total=int(input("ingrese el resultado total: "))
numero_sumado=int(input("ingrese el numero sumado: "))

# Despejamos la variable: x = 21 - 9
numero_incognito = resultado_total - numero_sumado

print(f"El número buscado es: {numero_incognito}")
```

#### Problema 2
Edades de Pedro y AnaPaso 1: Análisis y resolución matemáticaDefinir las variables: * Edad de Ana: $x$ (es nuestra referencia base).Edad de Pedro: $2x$ (porque el problema dice que es el "doble").Plantear la ecuación: * La suma de ambas edades es 36:$$x + 2x = 36$$Simplificar y despejar:Sumamos las términos semejantes: $3x = 36$Para dejar la $x$ sola, el 3 que está multiplicando pasa dividiendo:$x = \frac{36}{3}$Resolver:$x = 12$ (Edad de Ana)Pedro tiene el doble: $2 \times 12 = 24$Solución: Ana tiene 12 años y Pedro tiene 24 años.

```python
# Problema 2: Edad de Pedro (2x) + Edad de Ana (x) = 36
# suma_edades = 36
suma_edades=int(input("ingrese la suma de las edades: "))

# La ecuación simplificada es 3x = 36, por lo tanto x = 36 / 3
edad_ana = suma_edades / 3
edad_pedro = edad_ana * 2

print(f"La edad de Ana es: {int(edad_ana)} años")
print(f"La edad de Pedro es: {int(edad_pedro)} años")
```
