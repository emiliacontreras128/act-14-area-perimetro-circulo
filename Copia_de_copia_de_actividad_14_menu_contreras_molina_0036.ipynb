* 📘 Redacción: Problema y Fundamentos Matemáticos
En la geometría elemental, el círculo y la esfera son figuras esenciales para comprender cómo se mide el espacio en dos y tres dimensiones. El objetivo de este programa es ofrecer una herramienta interactiva y sencilla que permita calcular tres propiedades geométricas básicas a partir de un único dato de entrada: el radio (r).
* Redaccion congruente del problema
r = radio de la figura (siempre mayor a 0 en contextos reales, aunque el programa no validará este dato por solicitud).
π
π (pi) ≈ 3.14159, constante matemática fundamental en geometría circular.
* El programa está estructurado de forma modular y procedural (sin programación orientada a objetos), ideal para estudiantes de preparatoria. Utiliza un bucle while para mantener el menú activo hasta que el usuario decida salir, funciones independientes para cada cálculo, y estructuras de control if/elif/else para dirigir el flujo. Cada función solicita el radio internamente, realiza el cálculo con la constante math.pi de Python y muestra el resultado de forma clara y amigable.

formulas a utilizar
1. Area del circulo = A=π⋅r
2. perimetro del circulo = P=2⋅π⋅r
3. volumen de la esfera = P=2⋅π⋅r

```python
# ==========================================================
# 📐 Programa Interactivo de Geometría Básica
# Elaborado por ✅Contreras Molina
# Nivel: Principiante | Sin POO | Sin validación de entrada
# ==========================================================
import math

def menu():
    """Muestra las opciones y devuelve la elección del usuario."""
    print("\n" + "✨" * 25)
    print("📐 MENÚ DE CÁLCULOS GEOMÉTRICOS 📐")
    print("✨" * 25)
    print("1️⃣  Calcular área del círculo")
    print("2️⃣  Calcular perímetro del círculo")
    print("3️⃣  Calcular volumen de la esfera")
    print("4️⃣  Salir del programa")
    print("✨" * 25)
    opcion = int(input("👉 Elige una opción (1-4): "))
    return opcion

def areacirculo():
    """Calcula el área de un círculo: A = π * r²"""
    print("\n🔵 Cálculo del Área del Círculo")
    radio = float(input("📏 Ingresa el radio: "))
    area = math.pi * radio ** 2
    print(f"✅ El área es: {area:.2f} unidades²")

def perimetrocirculo():
    """Calcula el perímetro de un círculo: P = 2 * π * r"""
    print("\n🟡 Cálculo del Perímetro del Círculo")
    radio = float(input("📏 Ingresa el radio: "))
    perimetro = 2 * math.pi * radio
    print(f"✅ El perímetro es: {perimetro:.2f} unidades")

def volumenesfera():
    """Calcula el volumen de una esfera: V = (4/3) * π * r³"""
    print("\n🌍 Cálculo del Volumen de la Esfera")
    radio = float(input("📏 Ingresa el radio: "))
    volumen = (4/3) * math.pi * radio ** 3
    print(f"✅ El volumen es: {volumen:.2f} unidades³")

# ==========================================
# 🔁 Bucle principal de interacción
# ==========================================
print("🚀 ¡Bienvenido al Programa de Geometría Básica! 🚀")
while True:
    opcion = menu()

    if opcion == 1:
        areacirculo()
    elif opcion == 2:
        perimetrocirculo()
    elif opcion == 3:
        volumenesfera()
    elif opcion == 4:
        print("\n👋 ¡Gracias por usar el programa! Hasta pronto 🌟")
        break
    else:
        print("⚠️ Opción no válida. Por favor, selecciona un número del 1 al 4.")
```
